### ** Resizing an image size using different interpolation functions**

**1. ** Resize the lena or vibot image using nearest, bilinear and bicubic function.

You can define your function which accept the model and resize size from the user or just run a script.

```python
def resizeimage(image,sizeofimage,typeofimage):
    resizemyimage = misc.imresize(image,sizeofimage,interp=typeofimage)
    return resizemyimage
```

```python
#plt.figure
#plt.imshow(im_resize, cmap=plt.cm.gray)
#plt.show()  


Path = './images/'
im_ = 'vibot-color.jpg'
vibotcolor = io.imread(Path.__add__(im_))

plt.figure
plt.imshow(vibotcolor)
plt.title('The Vibot Image is Displayed')
plt.show()

# Use the nearest neighbour Interpolation
nearest = resizeimage(vibotcolor,(300,300),'nearest')
plt.figure
plt.imshow(nearest)
plt.title('Apply Nearest Neigbour on Vibot Image')
plt.show()

#  Use the bicubic Interpolation
bicubic = resizeimage(vibotcolor,(300,300),'bicubic')
plt.figure
plt.imshow(bicubic)
plt.title('Apply Bicubic Interpolation on Vibot Image')
plt.show()

#  Use the Bilinear Interpolation
bilinear = resizeimage(vibotcolor,(300,300),'bilinear')
plt.figure
plt.imshow(bilinear)
plt.title('Apply Bilinear Interpolation on Vibot Image')
plt.show()
```

**2. ** Similar to the function of `my_blending.m` that you saw before define your function of blending.

Blend the lena image and vibot image using simple, alpha and pyramid method.   

Pyramid blending method: 
1. Create a mask the same size as your input images 
2. Create Laplacian pyramid of the input images 
    2.1. Create Gaussian pyramids of the input images
    2.2. Create Laplacian pyramids using the Gaussian pyramids
3. Create Gaussian pyramid for the region mask
4. Blend the two laplacian pyramids using the pyramid of the mask as weigh 
    $$ BL = L_1 * (GM) + L_2 *(1 - GM)$$
5. Colapse the pyramid to get the final image 
    $$rescale(rescale(rescale(BL_3)+BL_2)+BL_1)+BL_0$$

Note : For pyramid blending approach, you can use:
    * transform.rescale for upscaling your images
    * transform.pyramid_gaussian for building your pyramids
    * The output of transform.pyramid_gaussian is a type generator and you need to convert it to a list :
    
        b = [a for a in Gaussian_pyramid]


```python
def samblending(image1,image2,btype,samalpha='0.5'):
    
    if btype == 'alpha':
        im = Image.blend(image1,image2,samalpha)
        return im
    
    elif btype == 'pyramid':
        a = np.matrix(np.zeros((512,256)))
        b = np.matrix(np.ones((512,256)))
        mask = np.concatenate((a,b),axis=1)
        
        gaussiansampyramid_image1 = transform.pyramid_gaussian(image1,max_layer=3)
        gaussiansampyramid_image1 = [a for a in gaussiansampyramid_image1]
        gaussiansampyramid_image2 = transform.pyramid_gaussian(image2,max_layer=3)
        gaussiansampyramid_image2 = [a for a in gaussiansampyramid_image2]
      
        laplaciansampyramid_image1 = []
        laplaciansampyramid_image2 = []
        
        for i in range(len(gaussiansampyramid_image1)-1) :
            laplaciansampyramid_image1.append(gaussiansampyramid_image1[i] - transform.rescale((gaussiansampyramid_image1[i+1]),2.0))
            laplaciansampyramid_image2.append(gaussiansampyramid_image2[i] - transform.rescale((gaussiansampyramid_image2[i+1]),2.0))
            
        laplaciansampyramid_image1.append(gaussiansampyramid_image1[3])
        laplaciansampyramid_image2.append(gaussiansampyramid_image2[3])
        
        gaussiansampyramid_mask = transform.pyramid_gaussian(mask,max_layer=3)
        gaussiansampyramid_mask = [a for a in gaussiansampyramid_mask]
     
        blendpyramid = []
        for i in range(len(gaussiansampyramid_mask)):
            blendpyramid.append(laplaciansampyramid_image1[i]* gaussiansampyramid_mask[i] + laplaciansampyramid_image2[i]
                                    *(1-gaussiansampyramid_mask[i]))
            
        blendpyramid_final = transform.rescale((transform.rescale((transform.rescale(blendpyramid[3],2.0)+
                                                                   blendpyramid[2]),2)+blendpyramid[1]),2.0)+blendpyramid[0]
        return blendpyramid_final
    

```
